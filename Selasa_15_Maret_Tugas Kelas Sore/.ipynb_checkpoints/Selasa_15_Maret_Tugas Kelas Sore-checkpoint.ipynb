```python
# Importing Data processing libraries
import numpy as np 
import pandas as pd 
pd.plotting.register_matplotlib_converters()

# Plotting Libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
```

```python
f,axes=plt.subplots(ncols=2, nrows=4, figsize=(10,10), sharex=True)

column=['Age', 'CustomerId', 'Balance', 'Exited']
for i in range(4):
    
    #Ploting the boxplot
    sns.boxplot(x = eda[column[i]], ax=axes[i][0], color='green')
    
    # Ploting the KDE
    sns.histplot(data=eda[column[i]], ax=axes[i][1],  color="white", kde=True, stat="density", linewidth=0)  
    
    plt.tight_layout()
```

```python
eda["Surname"].unique()
```

```python
# Dividing data
kelas2 = eda.loc[eda['Surname'] == "Hargrave"]
kelas3 = eda.loc[eda['Surname'] == "Onio"]
kelas4 = eda.loc[eda['Surname'] == "Kashiwagi"]
```

```python
kelas4.head
```

```python
eda.describe()
```

```python
f, axes = plt.subplots(ncols = 1,figsize=(5, 5), sharex=True)
sns.despine(left=True)


sns.histplot(data=kelas2["CustomerId"],label="Hargrave",color='r',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data=kelas3["CustomerId"],label="Onio",kde=True,color='b',linewidth=2,alpha=0.3)
sns.histplot(data=kelas4["CustomerId"],label="Kashiwagi",kde=True,color='g',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```

```python
f, axes = plt.subplots(ncols = 1,figsize=(5, 5), sharex=True)
sns.despine(left=True)

sns.histplot(data=kelas2["CreditScore"],label="Hargrave",color='c',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data=kelas3["CreditScore"],label="Onio",kde=True,color='orange',linewidth=2,alpha=0.3)
sns.histplot(data=kelas4["CreditScore"],label="Kashiwagi",kde=True,color='g',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```

```python
f, axes = plt.subplots(ncols = 1,figsize=(5, 5), sharex=True)
sns.despine(left=True)

sns.histplot(data=kelas2["EstimatedSalary"],label="Hargrave",color='c',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data=kelas3["EstimatedSalary"],label="Onio",kde=True,color='orange',linewidth=2,alpha=0.3)
sns.histplot(data=kelas4["EstimatedSalary"],label="Kashiwagi",kde=True,color='g',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```

```python
sns.pairplot(eda, hue="NumOfProducts", diag_kind="hist")
plt.show()
```

```python
f, axes = plt.subplots(ncols = 2, figsize=(10, 5), sharex=True)
sns.despine(left=True)

sns.swarmplot(x=eda['Age'], y=eda['Tenure'], ax=axes[0])

sns.boxplot(x=eda['Age'], y=eda['Tenure'], ax=axes[1])

plt.tight_layout()
```

```python
f, axes = plt.subplots(ncols = 2, figsize=(10, 5), sharex=True)
sns.despine(left=True)

sns.swarmplot(x=eda['CreditScore'], y=eda['Exited'], ax=axes[0])

sns.boxplot(x=eda['CreditScore'], y=eda['Exited'], ax=axes[1])

plt.tight_layout()
```

```python
f, axes = plt.subplots(ncols = 2, figsize=(10, 5), sharex=True)
sns.despine(left=True)

sns.swarmplot(x=eda['HasCrCard'], y=eda['EstimatedSalary'], ax=axes[0] ,size=4.5)

sns.boxplot(x=eda['HasCrCard'], y=eda['EstimatedSalary'], ax=axes[1])

plt.tight_layout()
```
